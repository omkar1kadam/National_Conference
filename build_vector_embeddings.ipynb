```python
from sentence_transformers import SentenceTransformer

# Load BGE Large model
model = SentenceTransformer("BAAI/bge-large-en")

print("Model loaded successfully!")
```

```python
word = "omkar"

embedding = model.encode(word)

print("Word:", word)
print("Embedding vector length:", len(embedding))
print("Embedding vector:\n", embedding)
```

```python
sentence = "Omkar is building a blockchain-based environmental monitoring system using ESP32 sensors and machine learning models."

embedding_sentence = model.encode(sentence)

print("Sentence:", sentence)
print("Embedding vector length:", len(embedding_sentence))
print("Embedding vector:\n", embedding_sentence)
```
